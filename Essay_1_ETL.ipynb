```python
import requests
import pandas as pd
```

```python
api_url = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"
```

**1. Buatlah API request untuk mendapatkan data return API berupa JSON**

```python
# Lakukan GET request ke API
response = requests.get(api_url)
```

```python
data = response.json().get('data')

# Transformasi data menjadi DataFrame
df = pd.DataFrame(data)

# Tampilkan DataFrame
print(df.head())
```

**(2) Ambil value dari key ‘data’ dalam JSON, dan transformasi menjadi sebuah DataFrame..**

```python
# Tampilkan DataFrame dalam bentuk tabel
display(df.head())
```

**(3) Transformasi tipe data kolom ‘isoDate’ menjadi tipe data datetime text**

```python
if response.status_code == 200:
    data = response.json().get('data')

    # Transformasi data menjadi DataFrame
    df = pd.DataFrame(data)

    # Transformasi kolom 'isoDate' menjadi tipe data datetime
    df['isoDate'] = pd.to_datetime(df['isoDate'])

     # Tampilkan DataFrame dalam bentuk tabel
    display(df.head())

else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
# Tampilkan tipe data setelah transformasi
display(df.dtypes)
```

**4. Aggregasi data berdasarkan kolom ‘creator’, serta urutkan berdasarkan jumlah berita terbanyak.**

```python
df['creator'] = 'cnn indonesia teknologi'

# Tampilkan DataFrame dalam bentuk tabel
display(df.head())
```

```python
# Agregasi data berdasarkan kolom 'creator' dan hitung jumlah berita
aggregasi_df = df.groupby('creator').size().reset_index(name='total')

# Urutkan DataFrame berdasarkan jumlah berita terbanyak
aggregasi_df = aggregasi_df.sort_values(by='total', ascending=False)

# Tampilkan DataFrame hasil agregasi dan pengurutan
display(aggregasi_df)
```
